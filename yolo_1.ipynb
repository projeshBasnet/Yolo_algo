# YOLO OBJECT DETECTION

## INTRODUCTION

**Yolo** is the state of art object detection techinique which has ability to detect objects in real time with high precision, accuracy within short duration of time. It is one of the top object detection algorithm used till date with high performance speed because it uses priciple of **you only look once**. There are four versions of yolo till current date and its performance has been boosted on each versions.

<img src="https://miro.medium.com/max/640/0*WUpMWzNu_ymDyHPp.png" />

### WHY IT IS FAST?

It is very fast as compared to other object detection algorithm like **R-CNN**,**Fast-Rcnn** and others because it moves through one gird of image only once during entire process where as in other algorithm multiple iteration of a same portion of image is done to detect object. This makes Yolo much more faster.   

### PROCESS OF OBJECT DETECTION

Yolo makes a specific grid of size s*s over the image to be detected. Than it process one grid of image at one time and detect weather object is present is that grid or not. And similary it traverse through  all girds of images and process those grids. 
When the  object is detected, it uses the grid which consist of mid-point of that object as reference. And than it returns the cordinate of midpoint along with height and width,label of object,confidence value of the object.And using that values we can create bounding boxes around object and label it with confidence.

### HANDLE MULTIPLE BOUNDING BOXES USING INTERSECTION OVER UNION(IOU) 

We can have multiple bounding boxes to a object that is detected because multiple midpoints can be calcuated according to size and number of grid used for object detection. So, in order to pick out most optimum grid box, we can use principle of (IOU). As name suggests intersection over union, we calcuate the area of union and intersection of two boxes. And calcuate ratio of intersection by union. If the ratio value is less than some specified thersehold than it means that two boxes are not intersected properly so, we can ignore condition of multi box bounding else there is intersection and it should be considered.

### NONE MAX SUPRESSION

When the value of Intersection over union(IOU) is above threasehold margin than we have to consider the condition of multiple box bounding. But now, there will be a question for selection for a box. As we can have only one bounding box for one object. Thus, we use principle of none max supression. Here, the confidence value of two overlapping bounding boxes is compared and box having less confidence value for defing that object will be discarded. In this way we can have one single bounding box which can depict the object boundary properly.

```python
import cv2
import numpy as np
```

```python
net = cv2.dnn.readNet('yolov3.weights','/content/darknet/cfg/yolov3.cfg')
```

```python
classes = []
with open('/content/darknet/data/coco.names','r') as f:
  classes = f.read().splitlines()
```

```python
img = cv2.imread('/content/darknet/data/person.jpg')
height,width = img.shape[:2]
width
img.shape
```

```python
blob = cv2.dnn.blobFromImage(img,1/255,(416,416),(0,0,0),swapRB=True,crop=False)
blob.shape
```

```python
net.setInput(blob)

output_layers_names = net.getUnconnectedOutLayersNames()
output_layers_names

layerOutputs = net.forward(output_layers_names)



```

```python
def imshow(img):
  import matplotlib.pyplot as plt
  %matplotlib.inline
  plt.gcf()
  plt.imshow(img)
```

```python
boxes = []
confidences = []
class_ids = []

for output in layerOutputs:
  for detection in output:
    scores = detection[5:]
    class_id = np.argmax(scores)
    confidence = scores[class_id]
    if confidence > 0.5:
      center_x = int(detection[0]*width) # value is being multiplied by width because to rescale value to original unit
      center_y = int(detection[1]*height)
      w = int(detection[2]*width)
      h = int(detection[3]*height)

      x = int(center_x - w/2)
      y = int(center_y - h/2)

      boxes.append([x,y,w,h])
      confidences.append(float(confidence))
      class_ids.append(class_id)


```

```python
class_ids
```

```python
indexes = cv2.dnn.NMSBoxes(boxes,confidences,0.5,0.4)
indexes.flatten()
```

```python
font = cv2.FONT_HERSHEY_PLAIN
colors = np.random.uniform(0,255,size=(len(boxes),3))
```

```python
if len(indexes.flatten()>0):
  for i in indexes.flatten():
    x,y,w,h = boxes[i]
    label = str(classes[class_ids[i]])
    confidence = str(round(confidences[i],2))
    color = colors[i]
    cv2.rectangle(img,(x,y),(x+w,y+h),color,2)
    cv2.putText(img,label + " " + confidence,(x,y+20),font,2,(255,255,255),2)
    
```
